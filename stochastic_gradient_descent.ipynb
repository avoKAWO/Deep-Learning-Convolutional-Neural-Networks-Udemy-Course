# Stochastic Gradient Descent Algorithm

$L(w)=w^2-4w$ \
$\frac{dL}{dw}=2*w-4$

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
```

```python
w = np.arange(-10, 14, 0.1)
Loss = w**2 - 4*w
point = -5
```

```python
plt.plot(w, Loss)
plt.plot(point, point**2 -4*point, 'ro')
plt.title('Loss Function')
```

```python
def stochastic_gradient_descent(df = lambda w: 2 * w - 4, learning_rate = 0.01, initial_weight = -5, max_iters = 10000, precision = 0.000001):
  # iteration counter
  iters = 0

  # control of the next drop step value
  previous_step_size = 1

  weights = []
  while previous_step_size > precision and iters < max_iters:
    previous_weight = initial_weight
    initial_weight -= (learning_rate * df(previous_weight))
    previous_step_size = abs(initial_weight - previous_weight)
    iters += 1
    weights.append(initial_weight)
    print('Iteration: #{}: Current point: {}'.format(iters, initial_weight))
  print('Global minimum: {}'.format(initial_weight))
  return weights
```

```python
weights = stochastic_gradient_descent(learning_rate = 0.1)
```

```python
weights
```

```python
plt.figure(figsize=(8, 6))
plt.plot(w, Loss)
for weight in weights:
  plt.plot(weight, weight**2 -4*weight, 'ro')
```
